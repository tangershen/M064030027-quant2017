## Q1


```python
%matplotlib notebook
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = web.DataReader("TSLA", 'google', datetime(2016,1,1))
```

```python
df.head()
```

```python
df[["Close","Volume"]].plot(subplots= True , figsize=(10,8) ,color="B")
plt.legend(loc='best')
```
