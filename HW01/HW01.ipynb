## Q3. 請使用 python 將新聞中可能的關鍵詞萃取出來！ (Bonus)

提示：

1. 使用老師提供的 util.py 來擷取鉅亨網上特定新聞的內容。
2. 使用 jieba 套件來擷取關鍵詞。

```python
import util
import jieba.analyse
```

```python
news_text = util.get_news_article('https://news.cnyes.com/news/id/3956525')
news_text
```

```python
## 全模式
seg_list=jieba.cut(news_text,cut_all=True)
print('Full Mode:' + '/'.join(seg_list)  )

```

```python
##精確模式
seg_list=jieba.cut(news_text,cut_all=False)
print("Default Mode: " + '/'.join(seg_list))
```

```python
##默認精確模式
seg_list=jieba.cut(news_text)
print('/'.join(seg_list))
```

```python
import jieba.analyse
jieba.analyse.extract_tags(news_text, topK=100)
```
